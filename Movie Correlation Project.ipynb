* **There is high correlation between budget and gross**

```python
# Visualize the correlation

correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title("Correlation Matrix for Numeric Features")

plt.xlabel("Movie Features")

plt.ylabel("Movie Features")

plt.show()
```

```python
# Convert the columns that are objects to category type to include in the correlation matrix

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes


df_numerized.head()
```

```python
# Visualize the correlation matrix for all the features

correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title("Correlation Matrix for Numeric Features")

plt.xlabel("Movie Features")

plt.ylabel("Movie Features")

plt.show()
```

```python
# Show the correlation of all the features

df_numerized.corr()
```

```python
# Show the correlation pairs

correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
# Sort the correlation pairs

sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
# Show the pairs with high correlation

high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```

## Results

* Votes and budget have the highest correlation to gross earnings: My educated guess here was correct

* Company has low correlation to gross earnings: My educated guess was wrong as there is low correlation between company and gross earnings
